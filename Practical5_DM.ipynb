```python
from sklearn.datasets import load_iris
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score,plot_confusion_matrix,classification_report
```

```python
from sklearn.model_selection import train_test_split,KFold
```

```python
data=load_iris()
```

```python
X=data.data
```

```python
Y=data.target
```

```python
data
```

```python
X
```

```python
Y
```

```python
X.shape
```

```python
Y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y, test_size=0.25,random_state=100)
```

```python
print("Shape of X train:",x_train.shape)
print("Shape of Y train:",y_train.shape)
print("Shape of X test:",x_test.shape)
print("Shape of Y test:",y_test.shape)
```

```python
dtc=DecisionTreeClassifier(criterion="entropy")
```

```python
dtc.fit(x_train,y_train)
```

```python
prediction=dtc.predict(x_test)
```

```python
from sklearn import tree
```

```python
import matplotlib.pyplot as plt
```

```python
fig,ax1=plt.subplots()
fig.set_size_inches(15,5)

ax1.set_title("Decision Tree-Holdout")
plot_confusion_matrix(dtc,x_test,y_test,ax=ax1)
```

```python
print("-----------Holdout-------------")
ac1=accuracy_score(y_test,prediction)*100
print("Accuracy:",ac1,"%")
print("Classification Report:")
print(classification_report(y_test,prediction))
```
